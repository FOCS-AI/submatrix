```python
import folium

def visualize_geojson_on_map(geojson_path):
    """
    Visualizes GeoJSON file on a Folium map.

    Parameters:
    - geojson_path: path to the GeoJSON file

    Returns:
    - A folium.Map instance
    """

    # Initialize a base map. For demonstration purposes, I'm centering the map at [0, 0] with zoom level 2.
    # Adjust these parameters as needed for your GeoJSON data.
    m = folium.Map(location=[0, 0], zoom_start=2)

    # Add the GeoJSON data to the map
    folium.GeoJson(geojson_path).add_to(m)

    return m

if __name__ == "__main__":
    geojson_file_path = input("Enter the path to your GeoJSON file: ").strip()
    map_instance = visualize_geojson_on_map(geojson_file_path)

    # Save the visualization as an HTML file
    output_html = "geojson_visualization.html"
    map_instance.save(output_html)
    print(f"GeoJSON visualization saved as '{output_html}'")
```
